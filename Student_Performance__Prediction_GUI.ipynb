```
import numpy as np
from flask import Flask, request, jsonify, render_template
import pickle
from flask_ngrok import run_with_ngrok


app = Flask(__name__, template_folder='templates')
model = pickle.load(open('Student_Performance_Prediction_model.pkl', 'rb'))
run_with_ngrok(app)

@app.route("/")
def home():
    return render_template('index.html')

@app.route('/predict',methods=['POST'])
def predict():
    performance = {
        0 : "Excellent",
        1 : "Failure",
        2 : "Good",
        3 : "Poor",
        4 : "Satisfactory"
    }
    int_features = [int(x) for x in request.form.values()]
    
    final_features = np.array(int_features)
    final_features = np.expand_dims(final_features, axis=0)
    # final_features = pd.DataFrame(int_features)
    prediction = model.predict_classes(final_features,batch_size=32,verbose=0)

    output = round(prediction[0], 2)

    # return render_template('index.html', prediction_text='Student Grade will be {}-----{}'.format(performance[output],int_features))
    return render_template('index.html', prediction_text='Student Grade will be {}'.format(performance[output]))


if __name__ == '__main__':
    app.run()
```
